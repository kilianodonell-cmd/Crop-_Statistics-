```python
import pandas as pd
import re
from rapidfuzz import fuzz

# 1. DYNAMIC MASTER SETUP
master_list = sorted(df_BR_num["stratum"].dropna().unique())
master_df = pd.DataFrame({
    "matching_id": range(1, len(master_list) + 1),
    "master_stratum": master_list
})

def norm_strict(s):
    if pd.isna(s): return ""
    return re.sub(r"[^a-z0-9]", "", str(s).lower())

master_df["_clean"] = master_df["master_stratum"].apply(norm_strict)

# 2. SMART MATCHING ENGINE
def find_smart_matches(s):
    clean_s = norm_strict(s)
    if not clean_s: return []

    # Priority 1: Exact Match (after cleaning)
    # If '10earlypedi' matches '10earlypedi', stop there.
    for _, row in master_df.iterrows():
        if clean_s == row["_clean"]:
            return [row["matching_id"]]

    # Priority 2: Substring/Combo Match
    # Only if it's a long string like '7latealluvial 8earlyalluvial'
    found_ids = []
    for _, row in master_df.iterrows():
        # We check if the master string is a distinct part of the messy string
        if row["_clean"] in clean_s:
            found_ids.append(row["matching_id"])

    # Priority 3: Fuzzy Match (Typos) - Only if nothing found yet
    if not found_ids:
        for _, row in master_df.iterrows():
            if fuzz.ratio(clean_s, row["_clean"]) > 85:
                found_ids.append(row["matching_id"])

    return list(set(found_ids)) # Remove any duplicates

# 3. APPLY AND EXPLODE
df_RP["matching_ids"] = df_RP["stratum"].apply(find_smart_matches)
df_RP_final = df_RP.explode("matching_ids").dropna(subset=["matching_ids"])
df_RP_final = df_RP_final.rename(columns={"matching_ids": "matching_id"})

# 4. FINAL VIEW
final_view = df_RP_final.merge(master_df, on="matching_id", how="inner")
summary = final_view.drop_duplicates(subset=["master_stratum"])

print(f"Verified Unique Matches: {len(summary)} out of {len(master_list)}")
print(summary[["matching_id", "master_stratum", "stratum"]].sort_values("matching_id"))
```
